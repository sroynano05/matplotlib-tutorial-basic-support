## Ploting Multiple Labels

```python
year= range(2012,2020)
apples=[0.896,0.91,0.919,0.926,0.931,0.929,0.931,0.937]
oranges=[0.962,0.941,0.930,0.918,0.308,0.907,0.904,0.901]
```

```python
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('years')
plt.ylabel('yeild (tons per hector)');
```

## *Line styling and Colouring*

```python
plt.plot(year,apples,marker='o',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy')
plt.plot(year,oranges,marker='x',c='r',ls='--',lw=3,ms=10,alpha=0.5)

plt.xlabel('years')
plt.ylabel('yeild (tons per hector)')

plt.title("crop yeild in kanto")
plt.legend(['apples','oranges']);
```

```python
plt.plot(year,apples,'s-b')
plt.plot(year,oranges,'o--r')

plt.xlabel('years')
plt.ylabel('yeild (tons per hector)')

plt.title("crop yeild in kanto")
plt.legend(['apples','oranges']);
```

```python
plt.plot(year,oranges,'or')
plt.title('yeild of orange (tons per hectors)');
```

## *Plotting the multiple charts  in a grid*

```python
fig,axes =plt.subplots(2,3,figsize=(16,8));
plt.tight_layout(pad=4)

axes[0,0].plot(year,oranges,'x--r')
axes[0,0].plot(year,apples,'o--m')
axes[0,0].set_title('yield of West Bengal')
axes[0,0].legend(['apples','oranges'])

sns.scatterplot(data=flowers_df,y=flowers_df.sepal_length,x=flowers_df.sepal_width,hue=flowers_df.species,ax=axes[0,1])

axes[0,2].hist([flowers_df.sepal_length,flowers_df.sepal_width],bins=np.arange(2,5,0.28),stacked=True)

sns.barplot(data=tips_df,y=tips_df.total_bill,x= tips_df.day,hue=tips_df.sex,ax=axes[1,0])

sns.heatmap(data=flights_df,cmap='Reds',ax=axes[1,1])

axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([]);



```
